```python
def load_book(name: str) -> str: 
  with open(file=f"./{name}", mode="r", encoding="utf8") as file:
    text = file.read()
    return text
  
  
def clean_text(text: str) -> str:
  return text.lower().replace("\n", "")
```

```python
print(load_book("book.txt"))
```

```python
def get_max_count(doc, type: str) -> str:
  hash_map = {}
  max_value = 0
  max_token = ""
  for token in doc:
    if token.pos_ == type:
      hash_map[token.lemma_] = hash_map.get(token.lemma_, 0) + 1
  for token, count in hash_map.items():
    if count > max_value:
      max_value = count
      max_token = token
  print(hash_map)
  return {"token": max_token, "count": max_value}
```

```python
def calc_max_values(doc, types: list[str]) -> dict[str, int]:
  hash_map = {}
  for type in types:
    hash_map[type] = get_max_count(doc, type)
  return hash_map
```

```python
import spacy
nlp = spacy.load('ru_core_news_sm')

text = load_book("book.txt")

cleaned_text = clean_text(text)

doc = nlp(cleaned_text)

max_count_items = calc_max_values(doc, ["ADJ", "NOUN", "VERB"])

tokens = [token.lemma_ for token in doc]
unique_tokens = set(tokens)

unique_percent = len(unique_tokens) / len(tokens) * 100

print(max_count_items)
print(f"Unique Percents: {unique_percent}") 
```
